Now, we can look at some correlation between features, especially between the champion feature.

```
correlation = df.corr()
print(correlation['champion'].sort_values(ascending = False),'\n')
```

```
f , ax = plt.subplots(figsize = (14,12))

plt.title('Correlation of Numeric Features with champion',y=1,size=16)

sns.heatmap(correlation,square = True,  vmax=0.8)

k= 11
cols = correlation.nlargest(k,'champion')['champion'].index
print(cols)
cm = np.corrcoef(df[cols].values.T)
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```

```
sns.set()
columns = correlation.nlargest(k,'champion')['champion'].index
sns.pairplot(df[columns],size = 2 ,kind ='scatter',diag_kind='kde')
plt.show()
```
